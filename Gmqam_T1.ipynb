```python
complete_df = pd.read_csv("/content/data.csv")


#----------------------ANALISE 1- ESTADO X PRESTADORES DE SERVIÇOS-------------------

#criação de uma tabela que mostra a relação da quantidade de prestadores de serviços diferentes por estado do brasil
df1 = complete_df[['sigla_uf','prestador_servico_esgotamento_sanitario']].groupby(['sigla_uf']).count().sort_values(by='prestador_servico_esgotamento_sanitario', ascending=False).drop_duplicates()
display(df1)

print("\n\n\n\n")

#Medidas de centralidade
descricao_numerica = df1.describe()
dfdescrição = pd.DataFrame(descricao_numerica)
display(dfdescrição)
print("\n\n\n\n")

#Box Plot
# Criar box plot para todas as colunas numéricas
df1.boxplot()
plt.title('Box Plot da quantidade de prestadores de serviço por Estado')
plt.ylabel('Quantidade de prestadores de serviço')
plt.grid(True)
plt.show()


print("\n\n\n\n")

#tentativa de gerar um grafico de linhas para representar a relação. (não consegui criar um histograma)
plt.plot(df1, marker='o', linestyle='-')
plt.xlabel('Estados')
plt.ylabel('Q de prestadores de serviço de tratamento de água)')
plt.title("Prestadores de serviço/Estado")
plt.show()

```

```python
#----------------------ANALISE 2 QUANTIDADE DE AGUA LANÇADA NOS CORPOS HIDRICOS ------------------
#Analise seguinte leva em conta a quantidade de agua com e sem tratamento lançada nos corpos hidricos por Estado
#A tabela é mostrada em ordem decrescente em da quantidade agua total lançada nos corpos hidricos (SP lança mais)
df2 = complete_df[['sigla_uf','carga_lancada_total', 'carga_lancada_com_coleta_com_tratamento', 'carga_lancada_com_coleta_sem_tratamento']].groupby('sigla_uf').sum().sort_values(by='carga_lancada_total', ascending=False)
#display(df2)
print("\n\n\n\n")

#Da quantidade de água total lançada nos corpos hídricos qual porcentagem não é tratada
#Na mesma tabela é criada mais uma coluna que mostra a porcentagem de agua lançada sem tratamento (ultima coluna)
#Mostra que mesmo que sp seja oq mais lança agua nos corpos hidricos, não é o Estado que lança mais agua sem tratamento, esse Estado é MG
df2['porcentagem_sem_tratamento'] = ((df2['carga_lancada_com_coleta_sem_tratamento'] / df2['carga_lancada_total'])) * 100
display(df2)
print("\n\n\n\n")

print("\n\n\n\n")
#Gráfico para representar qual Estado lança mais agua sem tratamento
df3 = df2[['porcentagem_sem_tratamento']]
df3 = df3.sort_values(by='porcentagem_sem_tratamento', ascending=False)
display(df3)
plt.plot(df3, marker='o', linestyle='-')
plt.xlabel('Estados')
plt.ylabel('Q de prestadores de serviço de tratamento de água)')
plt.title("Lançamento de água sem tratamento em corpos Hídricos por Estado")
plt.show()
print("\n\n\n")


#Medidas de centralidade
print('Medidas de Centralidade')
descricao_numerica2 = df2.describe()
dfdescrição2 = pd.DataFrame(descricao_numerica2)
display(dfdescrição2)
print("\n\n\n\n")


#Box Plot
# Criar box plot para todas as colunas numéricas
df3.boxplot()
plt.title('Box Plot da quantidade de agua lançada nos corpos hídricos no Brasil')
plt.ylabel('Quantidade de agua lançada')
plt.grid(True)
plt.show()
```

Conceito necessario para a analise do DBO


Avaliação da Poluição: A DBO é uma medida importante para avaliar a poluição orgânica em corpos d'água e efluentes. Altos níveis de DBO indicam a presença de uma grande quantidade de matéria orgânica que pode levar à depleção de oxigênio na água, afetando negativamente a vida aquática.

Controle de Processos de Tratamento: Em estações de tratamento de esgoto, a DBO é usada para monitorar a eficácia dos processos de tratamento. O objetivo é reduzir a DBO do efluente final para níveis seguros antes de liberar a água tratada no meio ambiente.

Baixa DBO: Indica uma baixa quantidade de matéria orgânica biodegradável. Em águas limpas e tratadas, a DBO deve ser baixa.

Alta DBO: Indica uma alta quantidade de matéria orgânica biodegradável, que pode ser uma indicação de poluição. Águas residuais e esgoto geralmente têm DBO alta.



Solução Conjunta: Adotar uma abordagem integrada e combinada para tratar a DBO, que pode incluir vários métodos e estratégias.


Corpo Receptor Intermitente ou Efêmero: Considerar a variabilidade e desafios associados a corpos d'água que não têm fluxo constante.


Requer Solução Complementar: Implementar medidas adicionais ou complementares para lidar com a DBO de maneira mais eficaz.

```python
df4 = complete_df[['sigla_uf', 'necessidade_remocao_dbo']]
display(df4)

#contagem de quantos casos obtemos de cada conclusão de diminuição da dbo
contagem = df4['necessidade_remocao_dbo'].value_counts()
print(contagem)
# Converter a série de contagem em um DataFrame
df_contagem = contagem.reset_index()
df_contagem.columns = ['necessidade de remoção dbo', 'Contagem']
#gráfico de barras
df_contagem.plot(kind='bar', x='necessidade de remoção dbo', y='Contagem', legend=False)
plt.xlabel('necessidade de remoção dbo')
plt.ylabel('Contagem')
plt.title('Análise da diminuição da BDO no Brasil')
plt.show()

#conclusão: Os corpos de tratamento analisados mostram que a maior parte deles encontram-se na situação de necessidade
# de mais de 80% de diminuir a dbo, podendo indicar grande indice de poluição

display(df4)
```
